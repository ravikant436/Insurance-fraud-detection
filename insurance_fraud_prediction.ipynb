```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

**Load Filtered Dataset**

```
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/Interview Topics/Insurance Fraund Detection/filtered_insurance_file.csv")
df.head()
```

```
# Separate data from Target variable
X = df.loc[:, df.columns != 'fraud_reported']
y = df['fraud_reported']
```

**Rescaling Data**

```
from sklearn import preprocessing

scaled_X = preprocessing.normalize(X)
```

#Logistic Regression (with K fold Cross validation and hypertuning)

```
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
from sklearn.model_selection import KFold
from sklearn.model_selection import GridSearchCV
```

```
log_classifier = LogisticRegression()
grid={'C':10.0 **np.arange(-2,3),'penalty':['l1','l2']}
cv = KFold(n_splits=5, shuffle=False, random_state=None)
```

```
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(scaled_X, y, train_size=0.7)
```

```
clf = GridSearchCV(log_classifier, grid, cv=cv, n_jobs=-1, scoring='f1_macro')
clf.fit(X_train, y_train)
```

```
y_pred = clf.predict(X_test)
print(y_pred)
```

```
print(np.array(y_test.to_list()))
```

```
y_pred = clf.predict(X_test)
print(metrics.confusion_matrix(y_pred, y_test))
```
